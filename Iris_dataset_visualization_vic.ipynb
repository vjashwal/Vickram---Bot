```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
### Question 3:   Is there a relationship between sepal length and petal length?
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', palette='Set2')
plt.title('Relationship Between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal_length', palette='Set2')
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='species', y='petal_length', palette='Set3', ci='sd')
plt.title('Bar Plot of Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='species', y='sepal_width', palette='Set1', inner='quartile')
plt.title('Violin Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
sns.pairplot(df, hue='species', palette='husl', diag_kind='kde', height=2)
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Adjust title position
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
### Question 4:   How does the petal width vary across different Iris species?
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=iris_df)
plt.title('Variation of Petal Width Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_width', data=iris_df, ci=None)
plt.title('Mean Petal Width Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Mean Petal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df, palette='Set2', s=100)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal_width', data=iris_df, inner='quartile', palette='muted')
plt.title('Violin Plot: Petal Width Across Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
sns.pairplot(iris_df, hue='species', palette='Set2', diag_kind='kde')
plt.suptitle('Pairwise Relationships in Iris Dataset', y=1.02)
plt.show()
```

```python
### Question 6:  Can we distinguish between Iris species based on sepal and petal measurements?
sns.pairplot(iris_df, hue='species', diag_kind='kde', palette='Set2')
plt.suptitle('Pairwise Relationships Between Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='sepal_length', data=iris_df, ci=None, palette='Set2')
plt.title('Average Sepal Length Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df, palette='Set2', s=100)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_width', data=iris_df, palette='muted')
plt.title('Box Plot: Sepal Width Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='petal_length', data=iris_df, inner='quartile', palette='coolwarm')
plt.title('Violin Plot: Petal Length Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
sns.pairplot(iris_df, hue='species', diag_kind='kde', palette='Set1')
plt.suptitle('Pairwise Relationships Between Sepal and Petal Features', y=1.02)
plt.show()
```
