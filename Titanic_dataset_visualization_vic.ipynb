```python
# Load the Titanic dataset
url = '/content/titanic.csv'
data = pd.read_csv(url)
```

```python
### Question 1: What is the distribution of passenger ages on the Titanic?
# Plotting age distribution
sns.histplot(data['Age'], bins=30, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
### Question 2: How does the survival rate vary based on passenger class (Pclass)?
# Plotting survival rate vary based on passenger class
sns.countplot(x='Pclass', hue='Survived', data=data, palette='pastel')
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Calculate survival rate by class
survival_class = data.groupby('Pclass')['Survived'].mean() * 100
print("Survival Rate by Passenger Class:\n", survival_class)
```

```python
### Question 3:  Is there a relationship between the fare paid and the likelihood of survival?
# Plotting fare paid vs likelihood of survival
sns.boxplot(x='Survived', y='Fare', data=data, palette='cool')
plt.title('Fare vs Survival')
plt.show()
```

```python
### Inference:
# - Passengers who survived tended to have paid higher fares.
```

```python
### Question 4:   How does the survival rate differ between male and female passengers?
# Plotting survival rate differ between male and female by gender
sns.countplot(x='Survived', hue='Sex', data=data, palette='viridis')
plt.title('Survival by Gender')
plt.show()
```

```python
### Question 4:   How does the survival rate differ between male and female passengers?
# Plotting survival rate differ between male and female by gender
sns.countplot(x='Survived', hue='Sex', data=data, palette='viridis')
plt.title('Survival by Gender')
plt.show()


# Calculate survival rate differ between male and female by gender
survival_gender = data.groupby('Sex')['Survived'].mean() * 100
print("Survival Rate by Gender:\n", survival_gender)
```

```python
### What impact does the number of siblings/spouses (SibSp) or parents/children (Parch) aboard have on survival?
# Plotting survival by siblings/spouses or parents/children (Parch) aboard
sns.countplot(x='SibSp', hue='Survived', data=data, palette='Set2')
plt.title('Survival by Siblings/Spouses Aboard')
plt.show()

sns.countplot(x='Parch', hue='Survived', data=data, palette='Set1')
plt.title('Survival by Parents/Children Aboard')
plt.show()
```

```python
### Inference:
# - Passengers with 1 or 2 siblings/spouses aboard had higher survival rates.

```

```python
### Inference:
# - Passengers with 1 or 2 parents/children aboard had higher survival rates.
```
